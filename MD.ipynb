## Exploratory Data Analysis - Summary

### Dataset Overview
- 5000 baris, 18 kolom
- Tidak ada missing values dan duplikat
- 2 target: placement_status (Klasifikasi) dan salary_package_lpa (Regresi)

### Temuan Utama
1. **Class Imbalance**: placement_status tidak seimbang (~82% tidak placed, ~18% placed)
2. **Fitur Paling Berpengaruh**: cgpa, technical_skill_score, soft_skill_score, dan backlogs menunjukkan perbedaan distribusi yang signifikan antara mahasiswa yang placed dan tidak placed
3. **Backlogs**: korelasi negatif terkuat (-0.40) terhadap placement_status — semakin banyak backlogs, semakin kecil peluang ditempatkan
4. **Gender & Extracurricular**: distribusi seimbang, tidak ada bias gender dalam dataset
5. **Salary**: mahasiswa yang tidak placed memiliki salary 0, sehingga salary sangat berkorelasi dengan placement_status (0.92)
6. **Feature Engineering**: student_id akan di-drop karena tidak relevan sebagai fitur prediksi

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

df_model = df.drop(columns=['student_id'])

le = LabelEncoder()
df_model['gender'] = le.fit_transform(df_model['gender'])
df_model['extracurricular_activities'] = le.fit_transform(df_model['extracurricular_activities'])

X = df_model.drop(columns=['placement_status', 'salary_package_lpa'])
y_cls = df_model['placement_status']
y_reg = df_model['salary_package_lpa']

X_train, X_test, y_cls_train, y_cls_test = train_test_split(X, y_cls, test_size=0.2, random_state=42)
_, _, y_reg_train, y_reg_test = train_test_split(X, y_reg, test_size=0.2, random_state=42)

print("X_train:", X_train.shape)
print("X_test:", X_test.shape)
```

## Modeling - Klasifikasi

### Pemilihan Fitur
Fitur yang digunakan dipilih berdasarkan hasil EDA:
- **cgpa, technical_skill_score, soft_skill_score**: menunjukkan perbedaan distribusi yang signifikan antara mahasiswa placed dan tidak placed
- **backlogs**: korelasi negatif terkuat (-0.40) dengan placement_status
- **gender, extracurricular_activities**: diikutsertakan sebagai konteks demografis
- **score_avg, academic_avg**: fitur hasil feature engineering untuk merangkum performa skill dan akademik

Fitur **student_id** di-drop karena hanya identifier, tidak memiliki nilai prediktif.

### Pemilihan Algoritma
Tiga algoritma dipilih dengan pertimbangan berikut:
1. **Logistic Regression**: digunakan sebagai baseline model — sederhana, interpretable, cocok untuk klasifikasi biner
2. **Random Forest**: ensemble method berbasis decision tree yang robust terhadap noise dan mampu menangkap interaksi non-linear antar fitur
3. **Gradient Boosting**: boosting method yang membangun model secara sekuensial untuk memperbaiki kesalahan prediksi sebelumnya, umumnya menghasilkan performa tinggi

### Metrik Evaluasi
- **Accuracy**: proporsi prediksi benar dari keseluruhan data
- **F1 Score**: digunakan karena dataset imbalanced (~82% kelas 0), F1 lebih representatif dibanding accuracy saja

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report

cls_models = {
    'Logistic Regression': LogisticRegression(max_iter=1000, random_state=42),
    'Random Forest': RandomForestClassifier(n_estimators=100, random_state=42),
    'Gradient Boosting': GradientBoostingClassifier(n_estimators=100, random_state=42)
}

cls_results = {}
for name, model in cls_models.items():
    model.fit(X_train, y_cls_train)
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_cls_test, y_pred)
    f1 = f1_score(y_cls_test, y_pred)
    cls_results[name] = {'Accuracy': acc, 'F1 Score': f1}
    print(f"\n{name}")
    print(classification_report(y_cls_test, y_pred))

print("\n=== Ringkasan Klasifikasi ===")
print(pd.DataFrame(cls_results).T)
```

## Modeling - Regresi

### Pemilihan Algoritma
1. **Linear Regression**: baseline model — mengasumsikan hubungan linear antara fitur dan target salary
2. **Random Forest Regressor**: mampu menangkap pola non-linear, robust terhadap outlier
3. **Gradient Boosting Regressor**: performa tinggi untuk data tabular dengan optimasi bertahap

### Metrik Evaluasi
- **RMSE (Root Mean Squared Error)**: mengukur rata-rata besarnya kesalahan prediksi dalam satuan yang sama dengan target (LPA). Semakin kecil semakin baik
- **R² Score**: mengukur seberapa besar variasi target yang dapat dijelaskan oleh model. Nilai mendekati 1 berarti model sangat baik

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score

reg_models = {
    'Linear Regression': LinearRegression(),
    'Random Forest Regressor': RandomForestRegressor(n_estimators=100, random_state=42),
    'Gradient Boosting Regressor': GradientBoostingRegressor(n_estimators=100, random_state=42)
}

reg_results = {}
for name, model in reg_models.items():
    model.fit(X_train, y_reg_train)
    y_pred = model.predict(X_test)
    rmse = np.sqrt(mean_squared_error(y_reg_test, y_pred))
    r2 = r2_score(y_reg_test, y_pred)
    reg_results[name] = {'RMSE': rmse, 'R2 Score': r2}
    print(f"\n{name} — RMSE: {rmse:.4f} | R2: {r2:.4f}")

print("\n=== Ringkasan Regresi ===")
print(pd.DataFrame(reg_results).T)
```

```python
from sklearn.model_selection import cross_val_score

for name, model in cls_models.items():
    scores = cross_val_score(model, X, y_cls, cv=5, scoring='f1')
    print(f"{name} — CV F1: {scores.mean():.4f} ± {scores.std():.4f}")
```

```python
df_model['score_avg'] = (df_model['technical_skill_score'] + df_model['soft_skill_score']) / 2
df_model['academic_avg'] = (df_model['ssc_percentage'] + df_model['hsc_percentage'] + df_model['degree_percentage']) / 3

X = df_model.drop(columns=['placement_status', 'salary_package_lpa'])
y_cls = df_model['placement_status']
y_reg = df_model['salary_package_lpa']

X_train, X_test, y_cls_train, y_cls_test = train_test_split(X, y_cls, test_size=0.2, random_state=42)
_, _, y_reg_train, y_reg_test = train_test_split(X, y_reg, test_size=0.2, random_state=42)

print("Fitur setelah feature engineering:", X.columns.tolist())
print("Shape:", X.shape)
```

- **Gradient Boosting**: CV F1 0.9994 — performa terbaik dan paling konsisten, dipilih sebagai model terbaik untuk klasifikasi

- **Random Forest Regressor**: R² 0.84, RMSE 1.55 — performa terbaik, mampu menangkap pola non-linear dengan baik
- **Gradient Boosting Regressor**: R² 0.82, RMSE 1.63 — performa sangat baik dan kompetitif
